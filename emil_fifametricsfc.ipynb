```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# import data 
file = "data.csv"
```

```python
df = pd.read_csv(file)
#pd.set_option('display.max_columns', None)
df.head()
```

```python
df.isnull().sum().head()
```

```python
# DATA CLEANING

# Cleaning up numeric columns : wage, value, release clause 
df["Release Clause"]=df["Release Clause"].str.replace("M","000000")
df["Release Clause"]=df["Release Clause"].str.replace("K","000")
df["Release Clause"]=df["Release Clause"].str.replace("€","")
df["Release Clause"]=df["Release Clause"].str.replace(".","")
df["Release Clause"] = df["Release Clause"].astype(float)

```

```python
df["Wage"]=df["Wage"].str.replace("M","000000")
df["Wage"]=df["Wage"].str.replace("K","000")
df["Wage"]=df["Wage"].str.replace("€","")
df["Wage"]=df["Wage"].str.replace(".","")
df["Wage"] = df["Wage"].astype(int)
```

```python
df["Value"]=df["Value"].str.replace("M","000000")
df["Value"]=df["Value"].str.replace("K","000")
df["Value"]=df["Value"].str.replace("€","")
df["Value"]=df["Value"].str.replace(".","")
df["Value"] = df["Value"].astype(int)

```

```python
df.head()
```

```python
# Combining the positions
def player_positions(x):
    a = ['GK']
    b = ['LWB','RWB','LB','LCB','CB','RCB','RB']
    c = ['LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LDM', 'CDM', 'RDM']
    d = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW']
    if x in a:
        return 'Goalkeeper'
    elif x in b:
        return 'Defenders'
    elif x in c:
        return 'Midfielder'
    elif x in d:
        return 'Attacker'
    else:
        return 'NaN'
df['Simplified Position'] = df['Position'].apply(player_positions)
df.head()
```

```python
df = df.rename(columns={"Release Clause": "Release Clause(€)",
                       "Wage": "Wage(€)",
                       "Value": "Value(€)"})
df.head()
```

```python
newdf= df[["Nationality","Value(€)"]]
newdf.head()
```

```python
TTnat= newdf.groupby('Nationality').sum().sort_values(by='Value(€)',ascending =False).head(10).reset_index()
TTnat.set_index('Nationality', inplace=True)
TTnat
```

```python
ax = TTnat.plot(kind='barh', title ="Top 10 Nations by Player Value", figsize = (10,10), edgecolor='red')
ax.set_ylabel("Nationality", fontsize=12)
ax.set_xlabel("Value(€)", fontsize=12)
ax.get_legend().remove()
plt.savefig("Top 10 Nations by Player Value.png")
plt.show()


```

```python
newdf2= df[["Club","Value(€)"]]
newdf2.head()
```

```python
TTclub= newdf2.groupby('Club').sum().sort_values(by='Value(€)',ascending =False).head(10).reset_index()
TTclub.set_index('Club', inplace=True)
TTclub
```

```python
ax = TTclub.plot(kind='barh', title ="Top 10 Clubs by Player Value", figsize = (10,10), edgecolor='green')
ax.set_ylabel("Club", fontsize=12)
ax.set_xlabel("Value(€)", fontsize=12)
ax.get_legend().remove()
plt.savefig("Top 10 Clubs by Player Value.png")
plt.show()
```
